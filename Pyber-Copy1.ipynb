```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn
seaborn.set()
colors=["gold", "light sky blue", "coral"]
palette=seaborn.xkcd_palette(colors)
seaborn.set_palette(palette)
seaborn.set_style({"axes.facecolor":"darkgray"})
```

```python
#Import and read the CSV file
# reading city data
city_data_df=pd.read_csv("city_data.csv")
city_data_df.head()
```

```python
# reading ride data
ride_data_df=pd.read_csv("ride_data.csv")
ride_data_df.head()
```

```python
merge_df= pd.merge(ride_data_df, city_data_df,how="left", on="city")
merge_df.head()
```

```python
grouped_df=merge_df.groupby("city").mean()
grouped_df



```

```python
bubble_df= merge_df.join(grouped_df, on="city", rsuffix="_average")
#drop columns ride_id average and driver count average
bubble_df=bubble_df.drop(["ride_id_average", "driver_count_average"],axis=1)
bubble_df.head()

```

```python
ride_counts=pd.DataFrame(bubble_df["city"].value_counts())
ride_counts.columns=["Number of Rides"]
ride_counts.head()
```

```python
bubble_df= bubble_df.join(ride_counts, on="city")
bubble_df.head()
```

```python
urban_cities=bubble_df[bubble_df["type"]=="Urban"]
rural_cities=bubble_df[bubble_df["type"]=="Rural"]
suburban_cities=bubble_df[bubble_df["type"]=="Suburban"]
urban_cities.head()
```

```python
plt.scatter(urban_cities["Number of Rides"],urban_cities["fare_average"],s=urban_cities["driver_count"],label="urban")
plt.scatter(rural_cities["Number of Rides"],rural_cities["fare_average"],s=rural_cities["driver_count"],label="rural")
plt.scatter(suburban_cities["Number of Rides"],suburban_cities["fare_average"],s=suburban_cities["driver_count"],label="suburban")
plt.xlabel("Number of Rides")
plt.ylabel("Average Fare ($)")
plt.legend()
plt.show()
            
```

```python
total_fares=bubble_df["fare"].sum()
urban_fares=urban_cities["fare"].sum()
suburban_fares=suburban_cities["fare"].sum()
rural_fares=rural_cities["fare"].sum()
total_fares
```

```python
urban_fares_percent=urban_fares/total_fares
suburban_fares_percent=suburban_fares/total_fares
rural_fares_percent=rural_fares/total_fares
```

```python
plt.pie([urban_fares_percent, suburban_fares_percent, rural_fares_percent],labels=["Urban", "Suburban","Rural"],autopct='%.2f%%')
plt.title("Percentage of Fares by City Type")
plt.show()
```

```python
#Number of Drivers
number_drivers=city_data_df.groupby("type").sum()
number_drivers
```

```python
total_drivers=number_drivers["driver_count"].sum()
total_drivers
```

```python
percent_drivers=number_drivers/total_drivers
percent_drivers
```

```python
plt.pie([urban_fares_percent, suburban_fares_percent, rural_fares_percent],labels=["Urban", "Suburban","Rural"],autopct='%.2f%%')
plt.title("Percentage of Drivers by City Type")
plt.show()
```
